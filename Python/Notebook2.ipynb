### Dizionari

**I dizionari in Python** sono strutture dati fondamentali utilizzate per memorizzare **coppie di chiavi e valori**. Ogni chiave è associata a un valore, consentendo un rapido accesso ai dati tramite la chiave corrispondente². Ecco una breve spiegazione:

1. **Cos'è un dizionario in Python?**
   - Un dizionario in Python collega termini e significati tra loro.
   - È simile a una **lista associativa**, ma con chiavi univoche.
   - Le chiavi e i valori possono essere collegati, memorizzati e recuperati successivamente.

2. **Struttura e interrogazione del dizionario:**
   - Un dizionario è racchiuso tra parentesi graffe `{}`.
   - Le coppie chiave-valore sono separate da virgole.
   - Esempio:
     ```python
     country_capitals = {
         "Germania": "Berlino",
         "Canada": "Ottawa",
         "Inghilterra": "Londra"
     }
     ```
   - Per accedere ai valori, usiamo le chiavi tra parentesi quadre (ad esempio, `country_capitals["Germania"]` restituirà "Berlino").

3. **Modificare e rimuovere dal dizionario:**
   - Possiamo aggiungere nuove coppie chiave-valore o modificare quelle esistenti.
   - Per rimuovere una coppia, usiamo `del country_capitals["Germania"]`.


````python
# Creazione di un dizionario
D = {
    "Germania": "Berlino",
    "Canada": "Ottawa",
    "Inghilterra": "Londra"
}

# Stampa del dizionario
print(D)
# Output: {'Germania': 'Berlino', 'Canada': 'Ottawa', 'Inghilterra': 'Londra'}

#Stampa delle chiavi
print(D.keys())
#stampa dei valori
print(D.values())
#Stampa tipo
print(type(D))
#Stampa lunghezza
print(len(D))

# Accesso ai valori tramite chiavi
print(D["Germania"])  # Output: Berlino
print(D["Inghilterra"])  # Output: Londra

# Aggiunta di un elemento al dizionario
D["Francia"] = "Parigi"

# Stampa del dizionario aggiornato
print(D)
# Output: {'Germania': 'Berlino', 'Canada': 'Ottawa', 'Inghilterra': 'Londra'(, 'Francia': 'Parigi'}

#Elenca le chiavi
for k in D:
    print(k)
#Elenca i valori attraverso le chiavi
for k in D:
    print(D[k])

#Metodi predefiniti dict
dir(D)
# Creazione di un dizionario
dizionario = {"nome": "Mario", "età": 30, "città": "Roma"}

# Uso del metodo keys() per ottenere le chiavi
print("Chiavi del dizionario:", dizionario.keys())

# Uso del metodo values() per ottenere i valori
print("Valori del dizionario:", dizionario.values())

# Uso del metodo items() per ottenere le coppie chiave-valore
print("Coppie chiave-valore del dizionario:", dizionario.items())

# Uso del metodo get() per ottenere il valore di una chiave specifica
print("Valore della chiave 'nome':", dizionario.get("nome"))

# Uso del metodo pop() per rimuovere una voce dal dizionario
valore_rimosso = dizionario.pop("età")
print("Valore rimosso:", valore_rimosso)
print("Dizionario dopo la rimozione:", dizionario)

# Uso del metodo update() per aggiornare il dizionario
dizionario.update({"professione": "ingegnere"})
print("Dizionario dopo l'aggiornamento:", dizionario)
# Creazione di un dizionario
dizionario = {"nome": "Mario", "età": 30, "città": "Roma"}

# Uso del metodo clear() per rimuovere tutte le voci dal dizionario
dizionario_temp = dizionario.copy()
dizionario_temp.clear()
print("Dizionario dopo clear():", dizionario_temp)

# Uso del metodo copy() per creare una copia del dizionario
dizionario_copia = dizionario.copy()
print("Copia del dizionario:", dizionario_copia)

# Uso del metodo fromkeys() per creare un nuovo dizionario con chiavi da un iterabile
nuovo_dizionario = dict.fromkeys(["chiave1", "chiave2", "chiave3"], "valore")
print("Nuovo dizionario da fromkeys():", nuovo_dizionario)

# Uso del metodo popitem() per rimuovere e restituire una coppia chiave-valore
chiave, valore = dizionario.popitem()
print(f"Coppia chiave-valore rimossa con popitem(): {chiave} - {valore}")
print("Dizionario dopo popitem():", dizionario)

# Uso del metodo setdefault() per ottenere il valore di una chiave o impostare un valore predefinito se la chiave non esiste
valore = dizionario.setdefault("professione", "ingegnere")
print("Valore ottenuto con setdefault():", valore)
print("Dizionario dopo setdefault():", dizionario)

#dict comprehension
numeri = [1, 2, 3, 4, 5]
quadrati = {n: n**2 for n in numeri}
print(quadrati)  # Output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# metodi predefiniti sui dict
D={"a":100,"b":1000}
dir(D)
print(D)
#D.clear()
D.pop("a") # estrae il valore corrispondente
           # alla chiave "a" e lo restituisce
print(D)

D={"a":100,"b":1000}
D.popitem() # toglie un elemento dal dict
# UN DIZIONARIO NON E' ORDINATO!!!
print(D)

# aggiungere un nuovo item a un dizionario
D["c"]=10000
print(D)

print(len(D))
 
````
